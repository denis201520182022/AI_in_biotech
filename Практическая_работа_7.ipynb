# Практическая работа 7. Основы Numpy. Построение графиков

## Numpy

Типовой синтаксис для подключения numpy и matplotlib

```python
import numpy as np
from matplotlib import pyplot as plt
```

Получить массив numpy можно из обычного списка

```python
my_list = [e for e in range(1, 11)]
print(my_list)

my_float_array = np.array(my_list, 'float')
my_int_array = np.array(my_list, 'int')

print(my_float_array, type(my_float_array))
print(my_int_array, type(my_int_array))
```

В отличие от списков массиви numpy позволяют реализовывать векторно-матричные операции обновременно над всеми элементами массива или матрицы

```python
# создаем массив чисел ль 1 до 10 с шагом 0.1
x = np.arange(0, 10, 0.5)
print(x)
y = np.sin(x)
print(y)
```

```python
# Выполним построение простейшего графика

x = np.arange(0, 100, 0.1)
y = 3 * np.power(x, 5) * np.sin(3 * x + 2)
plt.figure(figsize=(12,8))
plt.plot(x, y)
```

```python
# Выполним построение простейшей стобчатой диаграммы

x = np.array([i for i in range(0, 11, 1)])
y = np.sin(x)
plt.figure(figsize=(12,8))
plt.bar(x, y)
```

```python
# Реализуем визуализацию точечных данных

x1 = np.random.normal(loc = 1.0, scale = 0.5, size=1000)
y1 = np.random.normal(loc = 1.0, scale = 0.5, size=1000)

x2 = np.random.normal(loc = 4.0, scale = 0.8, size=1000)
y2 = np.random.normal(loc = 4.0, scale = 0.8, size=1000) + np.sin(3*x2)

plt.figure(figsize=(12,8))
plt.scatter(x1, y1)
plt.scatter(x2, y2)
```

```python
# Построим график функции заданный параметрически

t=np.arange(0, 360, 0.2)

x=4*np.cos(4*t) + 6*np.cos(5*t)
y=7*np.sin(1.5*t) + 5*np.sin(7*t)


plt.figure(figsize=(12,8))
plt.plot(x,y)
```
